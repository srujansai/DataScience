```python
import pandas as pd
import numpy as np
import sklearn
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
%matplotlib inline
plt.style.use('ggplot')
```

## Load Data

```python
amesHousingData = pd.read_excel('AmesHousing.xls')
amesHousingData.head()
```

```python
amesHousingData.isna().sum()[amesHousingData.isna().sum() > 0]
```

```python
contiColumns = ['Lot Area','Total Bsmt SF','1st Flr SF','2nd Flr SF','Garage Area','Pool Area','Misc Val','Gr Liv Area']
plt.figure(figsize=(25,18))
for i, datCol in enumerate(contiColumns):
    plt.subplot(2, 4, i+1)
    sns.distplot(amesHousingData[datCol])
    
plt.show()
```

```python
plt.figure(figsize=(25,18))
for i, datCol in enumerate(contiColumns):
    plt.subplot(2, 4, i+1)
    sns.scatterplot(x=datCol, y='SalePrice', data=amesHousingData)
    
plt.show()
```
